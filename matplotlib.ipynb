```python
df = pd.read_csv("datasets/daraz_dataset.csv")
```

```python
category = df.groupby("Category")["Product"].count()
plt.figure(figsize = (8,8))
plt.bar(category.index , category.values , color = "purple" , edgecolor ="red")
plt.xlabel("Category" , fontsize = 20)
plt.ylabel("Count", fontsize = 20)
plt.grid(axis = "y", linestyle="--", alpha=0.6)
```

```python
plt.pie(category.values , labels = category.index, shadow = True, startangle = 90)
plt.title("Categorical Distribution of Products ")
```

```python
gender_products = df.groupby("Gender")["Product"].count()
plt.pie(gender_products.values, labels = gender_products.index)
```

```python
bine =[]
plt.hist(df["UnitPrice"], edgecolor="black" , alpha = 0.7)
plt.title("Distribution of data based on Unit Price")
plt.xlabel("Unit Price")
plt.ylabel("frequency")
```

```python
plt.scatter(df["Age"] , df ["TotalAmount"] , color ="red" , s=80)
plt.xlabel("Age")
plt.ylabel("Total Amount")
plt.title("Age VS the Total Amount")
plt.grid(True)
```
