<h2>Outiler detection Using IQR</h2>

```python
# Outlier Detection of course_fee
q1 = df["course_fee"].quantile(0.25)
q3 = df["course_fee"].quantile(0.75)

iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)
outlier = df[(df["course_fee"] < lower_bound) | (df["course_fee"] > upper_bound)]
outlier["course_fee"]

# Outlier Detection of attandence_percentage
q1 = df["attendance_percent"].quantile(0.25)
q3 = df["attendance_percent"].quantile(0.75)

iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)
outlier = df[(df["attendance_percent"] < lower_bound) | (df["attendance_percent"] > upper_bound)]
outlier["attendance_percent"]

# Outlier Detection of assignment_score
q1 = df["assignment_score"].quantile(0.25)
q3 = df["assignment_score"].quantile(0.75)

iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)
outlier = df[(df["assignment_score"] < lower_bound) | (df["assignment_score"] > upper_bound)]
outlier["assignment_score"]

# Outlier Detection of final_score
q1 = df["final_score"].quantile(0.25)
q3 = df["final_score"].quantile(0.75)

iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)
outlier = df[(df["final_score"] < lower_bound) | (df["final_score"] > upper_bound)]
outlier["final_score"]
```

<h2> Outlier Removal</h2> 

```python
#From course_fee
df = df[
    (df["course_fee"] >= lower_bound) &
    (df["course_fee"] <= upper_bound)
]

median_fee = df["course_fee"].median()

df.loc[
    (df["course_fee"] < lower_bound) |
    (df["course_fee"] > upper_bound),
    "course_fee"
] = median_fee

df["course_fee"] = df["course_fee"].clip(
    lower=lower_bound,
    upper=upper_bound
)


#From attandence_percentage
df = df[
    (df["attandence_percentage"] >= lower_bound) &
    (df["attandence_percentage"] <= upper_bound)
]

md = df["attandence_percentage"].median()

df.loc[
    (df["attandence_percentage"] < lower_bound) |
    (df["attandence_percentage"] > upper_bound),
    "attandence_percentage"
] = md

df["attandence_percentage"] = df["attandence_percentage"].clip(
    lower=lower_bound,
    upper=upper_bound
)


#From assignment_score
df = df[
    (df["assignment_score"] >= lower_bound) &
    (df["assignment_score"] <= upper_bound)
]

md = df["assignment_score"].median()

df.loc[
    (df["assignment_score"] < lower_bound) |
    (df["assignment_score"] > upper_bound),
    "assignment_score"
] = md

df["assignment_score"] = df["assignment_score"].clip(
    lower=lower_bound,
    upper=upper_bound
)



#From final_score
df = df[
    (df["final_score"] >= lower_bound) &
    (df["final_score"] <= upper_bound)
]

md = df["final_score"].median()

df.loc[
    (df["final_score"] < lower_bound) |
    (df["final_score"] > upper_bound),
    "final_score"
] = md

df["final_score"] = df["final_score"].clip(
    lower=lower_bound,
    upper=upper_bound
)
```

<h2> Outlier Detection using Z-core</h2>

```python

mean = df["course_fee"].mean()
std = df["course_fee"].std()
df["course_fee_z"] = (df["course_fee"] - mean) / std
outlier["course_fee"]


mean = df["attandence_percentage"].mean()
std = df["attandence_percentage"].std()
df["attendance_percentage_z"] = (df["attandence_percentage"] - mean) / std
outlier["attandence_percentage"]



mean = df["assignment_score"].mean()
std = df["assignment_score"].std()
df["assignment_score_z"] = (df["assignment_score"] - mean) / std
outlier["assignment_score"]



mean = df["final_score"].mean()
std = df["final_score"].std()
df["final_score_z"] = (df["final_score"] - mean) / std
outlier["final_score"]


outliers = df[
    (df["course_fee_z"].abs() > 3) |
    (df["assignment_score_z"].abs() > 3) |
    (df["attendance_percentage_z"].abs() > 3) |
    (df["final_score_z"].abs() > 3)
]
outliers
```
